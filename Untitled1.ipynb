```python
import pandas as pd
import numpy as np
```

```python
file1 = pd.read_csv("./Data/nasdaq_screener_1644649434293.csv")
file2 = pd.read_csv("./Data/nasdaq_screener_1644649444015.csv")
```

```python
file1.head() #this is NASDAQ
#NOW NASDAQ is divided into 3-4 tiers, with one of the tiers being ADR (for owning foreign stuff)
```

```python
file2.head() #this is NYSE
```

```python
#get nasdaq list 
nasdaq_stocks = file1.Symbol.tolist()
nyse_stocks = file2.Symbol.tolist()
```

```python
with open ("nasdaq_stock_list.txt", 'w') as file:
    for item in nasdaq_stocks:
        file.write(f'{item}\n')

with open("nyse_stocks_list.txt", 'w') as file:
    for item in nyse_stocks:
        file.write(f'{item}\n')

        
```
